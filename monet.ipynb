# Building the Network

## Discriminator

```python
lrelu = lambda x: tf.keras.activations.relu(x, alpha=0.2)
```

```python
d_input = tf.keras.Input(shape=(256,256,3))
```

```python
x = tf.keras.layers.Conv2D(64, 4, 2, activation=lrelu)(d_input)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(128, 4, 2, activation=lrelu)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(256, 4, 2, activation=lrelu)(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(512, 4, 2, activation=lrelu)(x)
x = tf.keras.layers.BatchNormalization()(x)
```

```python
x.shape
```

```python
d_output = tf.keras.layers.Conv2D(1, (7, 14), 7, activation=lrelu)(x)
```

```python
disc = tf.keras.Model(inputs=d_input, outputs=d_output, name="Discriminator")
```

```python
disc.summary()
```

## Generator 

```python
g_input = tf.keras.Input(shape=(256,256,3))
```

```python
x = tf.keras.layers.Conv2D(64, 4, 2, activation="relu")(g_input)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(128, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(256, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(512, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(512, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2D(512, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
```

```python
x = tf.keras.layers.Conv2DTranspose(512, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2DTranspose(512, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2DTranspose(512, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
#x = tf.keras.layers.ZeroPadding2D()(x)
x = tf.keras.layers.Conv2DTranspose(256, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
#x = tf.keras.layers.ZeroPadding2D()(x)
x = tf.keras.layers.Conv2DTranspose(128, 4, 2, activation="relu")(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Conv2DTranspose(3, 4, 2, activation="tanh")(x)
```

```python
g_output = tf.keras.layers.ZeroPadding2D()(x)
```

```python
gen = tf.keras.Model(inputs=g_input, outputs=g_output, name="Generator")
```

```python
gen.summary()
```
